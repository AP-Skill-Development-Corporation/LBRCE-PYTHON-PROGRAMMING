### File handling in python

* File is a collection of records

* File handling:
    * creating a file
    * opening a file
    * writing data into a file
    * Reading data from a file
    * append data to a file
    * closing a file
    

### Mode of operations
* 'r' - read mode,which is used when the file is only being read
* 'w' - write mode,used to write data into a file,overwrites the file data
* 'a' - append mode, used to add data to already existing file data
* r+' - read and write mode,used to perform both read and write operations

##### In- build functions
* open(path,mode)
* close()

```python
# creating a file and writing data into a file
f=open("book.txt",'w')
f.write("hello world,welcome to python programming")
f.close()
```

```python
f=open("book.txt",'w')
f.write("this is my sample file")
f.close()
```

```python
f=open("book.txt",'a') # appending data to a file
f.write(" hello everyone")
f.close()
```

#### read data from a file

* read() -> used to read entire data from a file
* readlines() -> return entire file lines in list format
* readline() -> to read a single line from a file

```python
f=open("book.txt",'a')
f.write("\n python programming")
f.close()
```

```python
# read entire data from a file

f=open("book.txt",'r')
data=f.read()
print(data)
f.close()
```

```python
# read file data line by line
f=open("book.txt",'r')
lines=f.readlines()
print(lines)
for i in lines:
    print(i)
```

```python
# read file data character by character
f=open("book.txt",'r')
line=f.readline()
print(line)
for ch in line:
    print(ch)
```

```python
f=open("student.txt",'w')
for i in range(1,61):
    f.write("student %f \n "%(i))
f.close()
```

### methods in file handling
* tell() -> to know the current location of the file pointer
* seek() -> to set the file pointer at a specific location

```python
f=open("book.txt",'r')
print(f.tell())
data=f.readline()
print(data)
print(f.tell())
f.close()
```

```python
f=open("book.txt",'r')
print(f.seek(4))
data=f.readline()
print(f.seek(12))
f.close()
```

```python
# write a program to print no of words in a file
f=open("book.txt",'r')
data=f.read()
words=data.split()
nw=len(words)
print("number of words : ",nw)
f.close()
```

```python
# write a program to print no of characters ,digits, special chars
#in a file
f=open("book.txt",'r')
data=f.read()
nc=dc=spc=0
for char in data:
    if char.isalpha():
        nc+=1
    elif char.isdigit():
        dc+=1
    else:
        spc+=1
print("no of characters: ",nc)
print("no of digits: ",dc)
print("no of special characters: ",spc)  
```

### Regular expressions:

* RegEx or  Regular expression is a sequence of characters that forms a search pattern
* In python ,we have in-built package called re ,to work with regular expressions

```python
import re
help(re)
```

```python
txt="python programming"
#re.match(pattern,test_string)
re.match("python",txt)
```

```python
print(re.match("welcome",txt))
```

```python
pattern='^s...a$'
txt="shiva"
result=re.match(pattern,txt)
if result:
    print("search successful")
else:
    print("search not successful")
```

```python
re.match('^k..g$','king')
```

```python
s="hello everyone hello"
re.search('hello',s)
```

```python
s="hello everyone hello"
print(re.search('win',s))
```

```python
s="hello everyone hello"
re.findall('hello',s)
```

##### Comprehension
* used to reduce the lines of code

* Syntax:
* [output-variable loop condition]

```python
# list comprehension

[i for i in range(1,11)]
```

```python
[i for i in range(2,51,2)]
```

```python
[i for i in range(1,51) if i%2==0]
```

```python
# print odd numbers b/w 100 to 200 using comprehension
[i for i in range(100,200) if i%2!=0]
```

```python
[i+j for i in range(1,100) for j in range(100,200)]
```

```python
# square of natural numbers in range of 1 to 10
```

```python
[i**2 for i in range(1,11)]
```

```python
# cube of natural numbers in range of 1 to 10
[i**3 for i in range(1,11)]
```

#### dictionary comprehension
* if we write our code inside a dictionary and output is in the dictionary format then it is called as dictionary comprehension

```python
# 1:1 , 2:4,3:9,4:16 ...
{i:i**2 for i in range(1,11)}
```

```python
state=['A.P','Telangana','Tamilnadu']
capital=['Amaravathi','Hyderabad','chennai']
{k:v for (k,v) in zip(state,capital)}
```

```python
# print the student name who is having maximum marks

d={'stu1': 100, 'stu2': 20, 'stu3': 300}
m=max(d.values())
for k,v in d.items():
    if v==m:
        print("maximum marks student name is: ",k)
```

```python
d={'stu1': 100, 'stu2': 20, 'stu3': 300}
m=min(d.values())
for k,v in d.items():
    if v==m:
        print("minimum marks student name is: ",k)
```

### functional programming

* lambda
* anonymous function
* lambda arg1,arg2:expression

```python
# addition of 2 numbers using lambda

x=lambda a,b:a+b
x(90,78)
```

```python
# square of a given number
sq=lambda a:a**2
sq(9)
```

```python
# map(function,iterable)
```

```python
list(map(sq,range(1,11)))
```

```python
# filter(function,iterable)
```

```python
def evenCheck(a):
    if a%2==0:
        return True
    else:
        return False   
```

```python
list(filter(evenCheck,[11,12,13,17,18,80,78]))
```
